```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Extraction

```python
df = pd.read_csv('urban_vs_environment_2.csv')
df.head()
```

```python
df.info()
```

# Transform

```python
# Create copy of df
urbanization_df = df.copy()
```

```python
# Drop unnecessary columns
cols_to_drop = ['Unit of Measure','Definition','Data Coverage','Calendar Year','Base Year','Source','Footnotes']
urbanization_df.drop(columns=cols_to_drop, inplace=True)
urbanization_df.head()
```

```python
# Create indicator codes
urbanization_df['Indicator'].unique()
```

```python
# Create indicator codes dictionary
indicator_codes_dict = {
    "Annual Mean Levels (μg/m³) of Fine Particulate Matter (e.g., PM2.5 and PM10) in Cities (population weighted), Urban": "Urban Fine Particulate Matter Levels",
    "Annual Mean Levels (Î¼g/mÂ³) of Fine Particulate Matter (e.g., PM2.5 and PM10) in Cities (population weighted), Urban": "Urban Fine Particulate Matter Levels",
    "Fossil-fuel subsidies (consumption and production) as a proportion of total GDP (%)": "Fossil-fuel Subsidies and GDP Proportion",
    "Urban population (% of total population)": "Urban Population",
    "Total Greenhouse Gas Emissions (t '000 CO₂ equivalent)": "Total Greenhouse Gas Emissions",
    "Temperature change with respect to a baseline climatology, corresponding to the period 1951-1980": "Temperature Change",
    "Deforestation Rate (average % change)": "Deforestation Rate",
    "Growth Rates in Population": "Growth Rates in Population",
    "Carbon Dioxide Emissions (t '000 CO₂ equivalent)": "Carbon Dioxide Emissions",
    "Population density": "Population Density"
}
# Rename values in the Indicator column; keep original if not in dictionary
urbanization_df["Indicator"] = urbanization_df["Indicator"].map(indicator_codes_dict).fillna(urbanization_df["Indicator"])
urbanization_df['Indicator'].unique()
```

```python
# Round off year column values to 2 decimal places
year_cols = [col for col in urbanization_df.columns if str(col).isdigit() and 2010 <= int(col) <= 2025]
urbanization_df[year_cols] = urbanization_df[year_cols].apply(pd.to_numeric, errors='coerce').round(2)
```

```python
# Fill NaN values with 0 for year columns
urbanization_df[year_cols] = urbanization_df[year_cols].fillna(0)
urbanization_df.head()
```
