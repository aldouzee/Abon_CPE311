# Carbon Dioxide Emissions Data

```python
import pandas as pd
```

```python
df = pd.read_csv('carbon_dioxide_emissions.csv')
```

```python
# Create df copy
co2_df = df.copy()
co2_df.head()
```

```python
# Remove unnecessary columns
co2_df.drop(columns=['Series Code', 'Country Code'], inplace=True)
co2_df.head()
```

```python
# Remove bracketed year tags from year columns (e.g., '2016 [YR2016]' -> '2016')
co2_df.columns = co2_df.columns.str.replace(r'\s*\[YR\d{4}\]', '', regex=True)
```

```python
# Rename Series Names to shorter names
series_name_dict = {
    'Carbon dioxide (CO2) net fluxes from LULUCF - Deforestation (Mt CO2e)': 'Deforestation CO2 Emissions (tonnes)',
    'Carbon dioxide (CO2) emissions from Industrial Processes (Mt CO2e)': 'Industrial CO2 Emissions (tonnes)',
    'Carbon dioxide (CO2) emissions from Building (Energy) (Mt CO2e)': 'Building Energy CO2 Emissions (tonnes)',
}
co2_df['Series Name'] = co2_df['Series Name'].map(series_name_dict)
```

```python
# Change df to long format
year_cols = [col for col in co2_df.columns if str(col).isdigit()]

co2_long_df = co2_df.melt(
    id_vars=['Series Name', 'Country Name'],
    value_vars=year_cols,
    var_name='Year',
    value_name='CO2 Value'
)

co2_long_df['Year'] = co2_long_df['Year'].astype(int)
co2_long_df['CO2 Value'] = pd.to_numeric(co2_long_df['CO2 Value'], errors='coerce')
co2_long_df = co2_long_df.dropna(subset=['CO2 Value'])
co2_long_df = co2_long_df.sort_values(['Country Name', 'Series Name', 'Year']).reset_index(drop=True)
```

```python
# Save normal format as CSV
co2_df.to_csv('co2_emissions_clean.csv', index=False)
```
