1. Create a Food class that defines the following:
* name of the food
* value of the food
* calories of the food

2. Create the following methods inside the Food class:
* A method that returns the value of the food
* A method that returns the cost of the food
* A method that calculates the density of the food (Value / Cost)
* A method that returns a string to display the name, value and calories of the food

```python
class Food(object):
    def __init__(self, n, v, w, h):
        # Make the variables private
        self.name = n
        self.value = v
        self.calories = w
        #add happiness criterion
        self.happy = h
    def getValue(self):
        return self.value
    def getCost(self):
        return self.calories
    def density(self):
        return self.getValue()/self.getCost()
    # return happiness level
    def getHappy(self):
        return self.happy
    def __str__(self):
        return self.name + ': <' + str(self.value)+ ', ' + str(self.calories) + ', ' + str(self.happy) + '>'
```

3. Create a buildMenu method that builds the name, value and calories of the food


```python
def buildMenu(names, values, calories, happy):
    menu = []
    for i in range(len(values)):
        menu.append(Food(names[i], values[i],calories[i],happy[i]))
    return menu
```

4. Create a method greedy to return total value and cost of added food based on the desired maximum cost

```python
def greedy(items, maxCost, keyFunction):
    """Assumes items a list, maxCost >= 0,         keyFunction maps elements of items to numbers"""
    itemsCopy = sorted(items, key = keyFunction,
                       reverse = True)
    result = []
    totalValue, totalCost, totalHappy = 0.0, 0.0, 0.0
    for i in range(len(itemsCopy)):
        if (totalCost+itemsCopy[i].getCost()) <= maxCost:
            result.append(itemsCopy[i])
            totalCost += itemsCopy[i].getCost()
            totalValue += itemsCopy[i].getValue()
            totalHappy += itemsCopy[i].getHappy()
    return (result, totalValue)
```

5. Create a testGreedy method to test the greedy method

```python
def testGreedy(items, constraint, keyFunction):
    taken, val = greedy(items, constraint, keyFunction)
    print('Total value of items taken =', val)
    for item in taken:
        print('   ', item)
```

```python
def testGreedys(foods, maxUnits):
    print('Use greedy by value to allocate', maxUnits,          'calories')
    testGreedy(foods, maxUnits, Food.getValue)
    print('\nUse greedy by cost to allocate', maxUnits,          'calories')
    testGreedy(foods, maxUnits, lambda x: 1/Food.getCost(x))
    print('\nUse greedy by density to allocate', maxUnits,          'calories')
    testGreedy(foods, maxUnits, Food.density)
    print('\nUse greedy by happiness to allocate', maxUnits,          'happiness level')
    testGreedy(foods, maxUnits, Food.getHappy)
```

9. Create method to use  Bruteforce algorithm instead of greedy algorithm

```python
def maxVal(toConsider, avail):
    """Assumes toConsider a list of items, avail a weight
       Returns a tuple of the total value of a solution to the
         0/1 knapsack problem and the items of that solution"""
    if toConsider == [] or avail == 0:
        result = (0, ())
    elif toConsider[0].getCost() > avail:
        #Explore right branch only
        result = maxVal(toConsider[1:], avail)
    else:
        nextItem = toConsider[0]
        #Explore left branch
        withVal, withToTake = maxVal(toConsider[1:],
                                     avail - nextItem.getCost())
        withVal += nextItem.getValue()
        #Explore right branch
        withoutVal, withoutToTake = maxVal(toConsider[1:], avail)
        #Choose better branch
        if withVal > withoutVal:
            result = (withVal, withToTake + (nextItem,))
        else:
            result = (withoutVal, withoutToTake)
    return result
```

```python
def testMaxVal(foods, maxUnits, printItems = True):
    print('Use search tree to allocate', maxUnits,
          'calories')
    val, taken = maxVal(foods, maxUnits)
    print('Total costs of foods taken =', val)
    if printItems:
        for item in taken:
            print('   ', item)
```

```python
names = ['wine', 'beer', 'pizza', 'burger', 'fries','cola', 'apple', 'donut', 'cake']
values = [89,90,95,100,90,79,50,10]
calories = [123,154,258,354,365,150,95,195]
# add happy levels of food items
happy = [50, 25, 70, 80, 90, 75, 35, 70, 80]
foods = buildMenu(names, values, calories, happy)
testMaxVal(foods, 2400)
```

### Problem: Computer Resource Allocation
This program is capable of finding combinations having the most software applications open while maintaining the minimal resource usage. The resources include percentages of:
* CPU usage,
* GPU usage, memory usage, and
* CPU temperature
The algorithms used in this program include the greedy algorithm and the brute force algorithm.

Create Software class

```python
class Software(object):
    def __init__(self, n, c, m, t):
        self.name = n
        self.cpu = c
        self.memory = m
        self.temp = t
    def getCPU(self):
        return self.cpu
    def getGPU(self):
        return self.gpu
    def getMemory(self):
        return self.memory
    def getTemp(self):
        return self.temp
    def __str__(self):
        return self.name + ': <' + str(self.cpu) + ', ' + str(self.memory) + ', ' + str(self.temp) + '>'
```

Show Software resources

```python
def showResources(names, cpu_percent, memory_percent, temps):
    software_list = []
    for i in range(len(names)):
        software_list.append(Software(names[i], cpu_percent[i],  memory_percent[i], temps[i]))
        # print all software & resource usage
        # print(f'{names[i]} | CPU: {cpu_percent[i]}%, GPU: {gpu_percent[i]}%, Memory: {memory_percent[i]}%, Temp: {temps[i]}%')
    return software_list
```

```python
names = ['Microsoft Edge', 'Visual Studio Code', 'Adobe Photoshop', 'Spotify', 'League of Legends', 'Solitaire', 'File Explorer']
cpu_percent = [10, 40, 60, 15, 65, 5, 5]
memory_percent = [25, 35, 40, 15, 50, 5, 10]
temps = [10, 25, 40, 5, 40, 0, 0]
softwares = showResources(names, cpu_percent, memory_percent, temps)
```

Greedy method

```python
def resourceGreedy(items, maxPercent, keyFunction):
    softwares_copy = sorted(items, key = keyFunction,
                       reverse = True)
    result = []
    totalCPU, totalMemory, totalTemp = 0.0, 0.0, 0.0
    for i in range(len(softwares_copy)):
        if (totalCPU + softwares_copy[i].getCPU()) <= maxPercent:
            result.append(softwares_copy[i])
            totalCPU += softwares_copy[i].getCPU()
            totalMemory += softwares_copy[i].getMemory()
            totalTemp += softwares_copy[i].getTemp()
    return (result, totalCPU)
```

```python
def greedyTest(items, constraint, keyFunction):
    taken, val = resourceGreedy(items, constraint, keyFunction)
    print('Total value taken =', val)
    for item in taken:
        print('   ', item)
```

```python
def greedysTest(softwares, maxUnits):
    print(f'Max CPU Usage: {maxUnits}%')
    greedyTest(softwares, maxUnits, Software.getCPU)

    print(f'Max Memory: {maxUnits}%')
    greedyTest(softwares, maxUnits, Software.getMemory)

    print(f'Max Temperature: {maxUnits}%')
    greedyTest(softwares, maxUnits, Software.getTemp)
```

Brute force method

```python
def minVal(toConsider, avail):
    if toConsider == [] or avail == 0:
        result = (0, ())
    elif toConsider[0].getCPU() > avail:
        #Explore right branch only
        result = minVal(toConsider[1:], avail)
    else:
        nextItem = toConsider[0]
        #Explore left branch
        withVal, withToTake = minVal(toConsider[1:], avail - nextItem.getCPU())
        withVal += nextItem.getCPU()
        #Explore right branch
        withoutVal, withoutToTake = minVal(toConsider[1:], avail)
        #Choose better branch
        if withVal > withoutVal:
            result = (withVal, withToTake + (nextItem,))
        else:
            result = (withoutVal, withoutToTake)
    return result
```

```python
def testMinVal(softwares, maxUnits, printItems = True):
    print(f'Use search tree to allocate {maxUnits}%')
    val, taken = minVal(softwares, maxUnits)
    print(f'Total usage: {val}%')
    if printItems:
        for item in taken:
            print('   ', item)
```
