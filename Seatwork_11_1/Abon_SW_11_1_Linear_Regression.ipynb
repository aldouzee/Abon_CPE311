```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Part 1

Suppose we had a linear regression model with one feature that learned the following bias and weights:

- b=1
- w1=3

Further, suppose the following feature values for a given example:

- x1=0

Compute the linear regression prediction for this particular example using python in Google Colab. Make sure to plot the solution. Provide an analysis for the output.

```python
# linear regression prediction function
def r_model(b, w, x):
  return b + (w * x)
```

```python
# solution
b0 = 1
w1 = 3
x1 = 0

y = r_model(b0, w1, x1)
print(f'Y = {y}')
```

```python
# plot
x_values = np.linspace(x1 - 5, x1 + 5, 50)
y_values = r_model(b0, w1, x_values)

plt.figure(figsize=(8, 6))
plt.plot(x_values, y_values, color='blue')
plt.scatter(x1, y, color='red', marker='o', s=100)
plt.annotate(f'Prediction: ({x1}, {y})', (x1, y), textcoords="offset points", xytext=(10,10), ha='center', color='red')
plt.title(f'Linear Regression Line: y = {b0} + {w1}x')
plt.grid(True)
plt.show()
```

### Analysis
The plot  represents the linear relationship of the model y = 1 + 3x. The blue line shows the lienar regression line or how y changes as x changes. The red point at (0, 1)  illustrates the prediction for x1 = 0, which is y=1.

## Part 2

What will be the regression prediction with the following X values? Give the output. You may plot the regression line for additional points and give analysis.

Weights and Bias
- b=1
- w1=3
- w2=−2
- w3=4

Additional Example A:
- x1=3
- x2=-1
- x3=2

Additional Example B:
- x1=10
- x2=0
- x3=-10

```python
# linear regression prediction function
def logic_r_model(b, w_values, x_values):
  w_values = np.array(w_values)
  x_values = np.array(x_values)
  return b + np.dot(w_values, x_values)
```

```python
# solution
b=1
w_values = (3,-2,4)
x_example_a = (3,1,2)
x_example_b = (10,0,10)

# example a
y_a = logic_r_model(b, w_values, x_example_a)
# example b
y_b = logic_r_model(b, w_values, x_example_b)

print(f"Y of Example A = {y_a}\nY of Example B = {y_b}")
```

```python
x1_range = np.linspace(-10, 15, 50)
b_val = b
w_val = w_values

# Example A solution
x2_a_const = x_example_a[1]
x3_a_const = x_example_a[2]
y_a_line = []
for x_val_1 in x1_range:
    y_a_line.append(logic_r_model(b_val, w_val, [x_val_1, x2_a_const, x3_a_const]))

# Example B solution
x2_b_const = x_example_b[1]
x3_b_const = x_example_b[2]
y_b_line = []
for x_val_1 in x1_range:
    y_b_line.append(logic_r_model(b_val, w_val, [x_val_1, x2_b_const, x3_b_const]))

# plot
plt.figure(figsize=(8, 6))
# example a regression line
plt.plot(x1_range, y_a_line, color='blue', label=f'Regression Line A (x2={x2_a_const}, x3={x3_a_const})')
plt.scatter(x_example_a[0], y_a, color='red', marker='o', s=100, label=f'Prediction A: ({x_example_a[0]}, {y_a})')
plt.annotate(f'A: ({x_example_a[0]}, {y_a})', (x_example_a[0], y_a), textcoords="offset points", xytext=(10,10), ha='center', color='red')

# example b regression line
plt.plot(x1_range, y_b_line, color='green', label=f'Regression Line B (x2={x2_b_const}, x3={x3_b_const})')
plt.scatter(x_example_b[0], y_b, color='purple', marker='o', s=100, label=f'Prediction B: ({x_example_b[0]}, {y_b})')
plt.annotate(f'B: ({x_example_b[0]}, {y_b})', (x_example_b[0], y_b), textcoords="offset points", xytext=(10,10), ha='center', color='purple')

plt.title('Regression Lines and Predictions for Examples A and B')
plt.xlabel('x1')
plt.ylabel('Predicted Y')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Analysis
The plot shows the logistic regressions of examples A and B. Example A shows that the relationship between x and y is (3,16). Example B shows that the relationship between x and y is (10,71).
