### Exercise 2

Change date column into datetime, volume column to integer

```python
faang_df = pd.read_csv("faang.csv", index_col=0)
```

```python
faang_df.dtypes
```

```python
faang_df["date"] = pd.to_datetime(faang_df["date"])
faang_df["date"].dtypes
```

```python
faang_df["volume"] = faang_df["volume"].astype(int)
faang_df["volume"].dtypes
```

Sort by date and ticker

```python
faang_df = faang_df.sort_values(by=["date", "ticker"])
faang_df
```

Find seven rows with highest volume

```python
faang_df_highest_volume = faang_df.sort_values(by="volume")
faang_df_highest_volume.head(7)
```

Use melt() to make completely long format

```python
faang_df = faang_df.melt(id_vars="date", value_vars=["open", "high", "low", "close", "volume"], var_name="price_point", value_name="stock_price", ignore_index=False)
faang_df
```

### Exercise 3

Install dependencies

```python
import requests
from selenium import webdriver
from selenium.webdriver.firefox.service import Service
from selenium.webdriver.firefox.options import Options
from webdriver_manager.firefox import GeckoDriverManager
import time
from bs4 import BeautifulSoup
```

```python
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64: x64)'
}
url = "https://hmocard.ph/hmo-accredited-hospitals-in-ncr/"
response = requests.get(url, headers=headers)
print(response.status_code)
print(response.text)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
print(soup.find('title').text)
```

```python
data = soup.find(class_="wp-block-table")
print(data.prettify())
```

```python
hospital_names = []
med_center_types = []
cities = []
regions = []
addresses = []
contact_numbers = []
```

```python
rows = soup.find('table').find_all('tr')
for row in rows:
    cells = row.find_all('td')
    if cells:
        hospital_name = cells[0].strip()
        hospital_names.append(hospital_name)
        
        med_center_type = cells[1].strip()
        med_center_types.append(med_center_type)

        city = cells[2].text.strip()
        cities.append(city)

        region = cells[3].text.strip()
        regions.append(region)

        address = cells[4].text.strip()
        addresses.append(address)

        contact_number = cells[5].text.strip()
        contact_numbers.append(contact_number)
```
