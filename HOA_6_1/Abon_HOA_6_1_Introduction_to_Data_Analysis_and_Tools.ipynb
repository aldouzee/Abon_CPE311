## Exercise 1

Run the given code below for exercises 1 and 2, perform the given tasks without using any Python modules.

```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

Using the data generated above, calculate the following statistics without importing anything from the statistics module in the
standard library (https://docs.python.org/3/library/statistics.html) and then confirm your results match up to those that are
obtained when using the statistics module (where possible):<br>
- Mean
- Median
- Mode (hint: check out the Counter in the collections module of the standard library at
https://docs.python.org/3/library/collections.html#collections.Counter)
- Sample variance
- Sample standard deviation

```python
salaries
```

```python
# Mean
salaries_mean = sum(salaries) / len(salaries)
print(f'Mean Salary: {salaries_mean}')
```

```python
# Median
salaries_median = salaries[len(salaries)//2]
print(f'Median Salary: {salaries_median}')
```

```python
# Mode
salaries_mode = max(salaries, key=lambda x: salaries.count(x))
salaries_mode
```

```python
# Sample variance
average = salaries_mean
len_minus_one = len(salaries)-1

x_mean_squared = []
for salary in salaries:
  temp = (salary - average)**2
  x_mean_squared.append(temp)

summation = sum(x_mean_squared)
salaries_variance = summation/len_minus_one

print(f'Salaries Sample Variance: {salaries_variance:.2f}')
```

```python
# Sample standard deviation
salaries_std = salaries_variance ** 0.5

print(f'Salaries Standard Deviation: {salaries_std:.2f}')
```

# Exercise 2

Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:<br>
- Range
- Coefficient of variation
- Interquartile range
- Quartile coefficient of dispersion

```python
# Range
salaries_range = max(salaries) - min(salaries)
print(f'Salaries Range: {salaries_range}')
```

```python
# Coefficient of variation
salaries_cov = salaries_std / salaries_mean
print(f'Salaries Coefficient of Variation: {salaries_cov}')
```

```python
# Interquartile range
salaries.sort()
q1 = .25 * (len(salaries)+1)
q3 = .75 * (len(salaries)+1)
salaries_iqr = q3 - q1

print(f'Salaries Interquartile Range: {salaries_iqr}')
```

```python
# Quartile coefficient of dispersion
salaries_qcd = (q3-q1)/(q3+q1)

print(f'Salaries Quartile Coefficient of Dispersion: {salaries_qcd}')
```
