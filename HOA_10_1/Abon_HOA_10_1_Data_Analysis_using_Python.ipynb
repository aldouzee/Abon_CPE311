# Transform

```python
# create df copy
food_sec_df = df.copy()
food_sec_df.head(5)
```

```python
# remove first row
food_sec_df = food_sec_df.drop(0)
```

```python
# remove unnecessary columns
""""
- Iso3
- StartDate
- EndDate
- Area Code
- Area Code (M49)
- Area
- Year Code
- Note
"""

cols_to_remove = ["Iso3", "StartDate", "EndDate", "Area Code", "Area Code (M49)", "Area", "Year Code", "Note"]
food_sec_df = food_sec_df.drop(columns=cols_to_remove)

food_sec_df.head(5)
```

```python
# checking null values and data formats
food_sec_df.info()
```

```python
# remove rows with unknown units and values
food_sec_df = food_sec_df[food_sec_df["Unit"].notna()]
food_sec_df = food_sec_df[food_sec_df["Value"].notna()]
```

```python
# convert data types
food_sec_df["Element Code"] = food_sec_df["Element Code"].astype(int)
```

```python
food_sec_df.info()
```

```python
# rearrange columns
food_sec_df = food_sec_df[["Year", "Item Code", "Item", "Element Code", "Element", "Unit", "Value", "Flag"]]
```

# Load

## Visualization Goal
The following visualizations should show the effectiveness of the Philippines' actions for SDG 2 - Zero Hunger.

## Food Security Indicator Values

### Indicators

21010 – Average dietary energy supply adequacy (percent) (3-year average)<br>
- Measures the adequacy of food energy supply relative to requirements.
- Questions: Is the population’s energy intake meeting the recommended threshold?

220001 – Dietary energy supply used in the estimation of prevalence of undernourishment (kcal/cap/day)<br>
- Daily per-capita calorie supply used in PoU calculations.
- Questions: What is the per-capita calorie supply used to estimate hunger?

22000 – Dietary energy supply used in the estimation of prevalence of undernourishment (kcal/cap/day) (3-year average)<br>
- 3-year average of the per-capita calorie supply metric.
- Questions: Is multi-year calorie supply stable or changing?

21012 – Share of dietary energy supply derived from cereals, roots and tubers (percent) (3-year average)<br>
- Reflects dietary diversity and staple reliance.
- Questions: Is the diet dominated by staples?

21013 – Average protein supply (g/cap/day) (3-year average)<br>
- Tracks protein availability in the national food supply.
- Questions: Are protein needs being met?

22013 – Gross domestic product per capita, PPP (constant 2017 international $)<br>
- Proxy for economic access to food.
- Questions: How does income influence affordability and access?

210041 – Prevalence of undernourishment (percent) (3-year average)<br>
- Proportion of population below minimum energy requirements.
- Questions: How widespread is chronic hunger?

210011 – Number of people undernourished (million) (3-year average)<br>
- Absolute scale of undernourishment.
- Questions: How many individuals are affected?

210091 – Prevalence of moderate or severe food insecurity in the total population (percent) (3-year average)<br>
- Share of population facing moderate or severe food insecurity.
- Questions: What portion of the population lacks reliable access to adequate food?

210081 – Number of moderately or severely food insecure people (million) (3-year average)<br>
- Absolute count of people facing moderate or severe food insecurity.
- Questions: What is the population-level burden (for planning/resources)?

These indicators capture the state of calorie adequacy, protein availability, dietary composition, prevalence and scale of undernourishment and food insecurity in the country.

```python
# create dictionary of item codes and names
items = {
    '21010': 'Average dietary energy supply adequacy (percent) (3-year average)',
    '220001': 'Dietary energy supply used in the estimation of prevalence of undernourishment (kcal/cap/day)',
    '22000': 'Dietary energy supply used in the estimation of prevalence of undernourishment (kcal/cap/day) (3-year average)',
    '21012': 'Share of dietary energy supply derived from cereals, roots and tubers (percent) (3-year average)',
    '21013': 'Average protein supply (g/cap/day) (3-year average)',
    '22013': 'Gross domestic product per capita, PPP, (constant 2017 international $)',
    '210041': 'Prevalence of undernourishment (percent) (3-year average)',
    '210011': 'Number of people undernourished (million) (3-year average)',
    '210091': 'Prevalence of moderate or severe food insecurity in the total population (percent) (3-year average)',
    '210081': 'Number of moderately or severely food insecure people (million) (3-year average)'
}

# setting figure cols and rows, color palette
n_cols = 3
n_rows = (len(items) + n_cols - 1) // n_cols
palette = sns.color_palette('tab10', n_colors=len(items))

# creating figure and axes
fig, axes = plt.subplots(n_rows, n_cols, figsize=(16, n_rows * 3))
axes = axes.flatten()

# plotting
for i, (code, label) in enumerate(items.items()):
    df_temp = food_sec_df[food_sec_df['Item Code'] == code].copy()
    if df_temp.empty:
        continue
    df_temp['Year'] = df_temp['Year'].astype(int)
    df_temp['Value'] = df_temp['Value'].astype(float)

    sns.barplot(df_temp, x='Year', y='Value', color=palette[i], ax=axes[i])
    axes[i].set_title(label, fontsize=7)
    axes[i].set_xlabel('Year', fontsize=7)
    axes[i].set_ylabel('Value ' + '('+ df_temp['Unit'].iloc[0] + ')', fontsize=7)
    axes[i].tick_params(axis='x', rotation=45, labelsize=7)
    axes[i].tick_params(axis='y', labelsize=7)

for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

plt.suptitle("Food Security Indicator Values")
plt.tight_layout()
plt.show()
```

### Analysis
The visualization mainly reflects all factors relating to the food security of the Philippines. It includes the yearly trends of food indicators like average dietary energy supply adequacy, average protein supply, gross domestic product per capita, and prevalence of undernourishment.

Increases are seen in the average dietary energy supply adequacy, average protein supply, and gross domestic product per capita are evident, especially starting from 2016 and so on. Moreover, a huge decrease is seen in the prevalence of undernourishment, especially starting from year 2016 and so on.

Overall, we can see positive results on different food indicators, proving the effectiveness of reaching SDG 2.

## Dietary Energy Supply vs. Prevalence of Undernourishment

```python
# prepare subset
nutritional_adequacy_df = food_sec_df[food_sec_df['Item Code'].isin(['21010', '210041'])].copy()

# plotting
plt.figure(figsize=(10, 5))
sns.lineplot(nutritional_adequacy_df, x='Year', y='Value', hue='Item')

plt.suptitle('Dietary Energy Supply vs. Prevalence of Undernourishment')
plt.xlabel('Year')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.legend(title='Legend', bbox_to_anchor=(1, 1))

plt.tight_layout()
plt.show()
```

### Analysis
This visualization clearly shows a strong inverse relationship between dietary energy supply and undernourishment in the Philippines. It demonstrates that increasing dietary energy supply has been a key driver in reducing undernourishment in the Philippines, though improvements in access and policy likely played a crucial role in accelerating progress after 2015.

## Economic Growth vs. Undernourishment

```python
# prepare subsets
gdp_vs_prevalence_df = food_sec_df[food_sec_df['Item Code'].isin(['22013', '210041'])].copy()
gdp_vs_prevalence_df['Year'] = gdp_vs_prevalence_df['Year'].astype(int)
gdp_vs_prevalence_df['Value'] = gdp_vs_prevalence_df['Value'].astype(float)

df_gdp = gdp_vs_prevalence_df[gdp_vs_prevalence_df['Item Code'] == '22013']
df_undernourished = gdp_vs_prevalence_df[gdp_vs_prevalence_df['Item Code'] == '210041']
```

```python
# plotting
fig, ax1 = plt.subplots(figsize=(12,5))
ax2 = ax1.twinx()

sns.lineplot(x='Year', y='Value', data=df_gdp, ax=ax1, color='green', label='GDP per capita')
sns.lineplot(x='Year', y='Value', data=df_undernourished, ax=ax2, color='orange', label='Prevalence of Undernourishment')

ax1.set_xlabel('Year')
ax1.set_ylabel('GDP per capita (constant 2017 $)', color='green')
ax2.set_ylabel('Prevalence of Undernourishment (%)', color='orange')

ax1.tick_params(axis='y', labelcolor='green')
ax2.tick_params(axis='y', labelcolor='orange')

ax1.set_xticks(gdp_vs_prevalence_df['Year'].unique())
plt.suptitle('GDP per capita vs. Prevalence of Undernourishment')
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')
plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

### Analysis
The visualization shows a strong inverse relationship between GDP per capita and undernourishment in the Philippines. This indicates that economic growth has also played a critical role in improving food security in the country. Even with a temporary dip in GDP around 2020, which is the COVID-19 pandemic period, undernourishment continued its downward trend, suggesting that policy and social interventions may have helped the decrease of undernourishment prevalence.

## Relationship Between Nutrition & Economic Status

```python
# prepare subsets
df_gdp = food_sec_df[food_sec_df['Item Code'] == '22013'][['Year', 'Value']].rename(columns={'Value': 'GDP per capita'})
df_energy = food_sec_df[food_sec_df['Item Code'] == '21010'][['Year', 'Value']].rename(columns={'Value': 'Dietary Energy supply'})
df_protein = food_sec_df[food_sec_df['Item Code'] == '21013'][['Year', 'Value']].rename(columns={'Value': 'Protein supply'})

df_merged = df_gdp.merge(df_energy, on='Year').merge(df_protein, on='Year')
```

```python
# find correlation
corr = df_merged[['GDP per capita', 'Dietary Energy supply', 'Protein supply']].corr()

# plotting
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

plt.xticks(rotation=45)
plt.suptitle('Correlation Between GDP and Nutritional Supply')
plt.show()
```

### Analysis
The visualization reveals a strong positive correlation between the Philippines' economic performance and its nutritional indicators.

There is a high correlation between GDP per capita and Protein supply, with 0.96. This suggests that as the economy grows, the population’s access to protein-rich foods increases. Dietary energy supply also shows a high relationship with GDP with 0.88, indicating that economic gains are a primary driver in meeting the country's dietary energy supply requirements. The high correlation between protein and energy supply reinforces that these indicators move together, with 0.84.

# Conclusion

The visualizations demonstrate a significant upward trend in the Philippines' food security indicators.

A high correlation between economic growth and nutritional improvement is seen, indicating that the national wealth remains the most effective driver for improving the nation's food security. The prevalence of undernourishment also decreased from 18% to under 5% in a short period of time.

Despite declining prevalence in undernourishment, the data show that moderate and severe food insecurity are still affecting over 30% of the population suggesting that the stability and affordability of that food supply remain vulnerable to external factors like crises and outbreaks.

Overall, the data indicates that the Philippines is creating a significant positive actions in achieving food security in the nation.
