## Exercise Part 4

```python
# Meteorite data
meteor_landings = pd.read_csv('../[00] Datasets/Meteorite_Landings.csv')
meteor_landings.head()
```

```python
# Update the year column to only contain the year
meteor_landings['year'] = meteor_landings['year'].str.slice(6, 10)
meteor_landings['year'] = pd.to_numeric(meteor_landings['year'], errors='coerce').astype('Int64')
```

```python
meteor_landings.head()
```

```python
# Pivot table: meteorite count and 95th percentile mass by fall type (2005-2009)
meteor_landings['mass (g)'] = pd.to_numeric(meteor_landings['mass (g)'], errors='coerce')
```

```python
filtered = meteor_landings[
    meteor_landings['year'].between(2005, 2009)
    & meteor_landings['fall'].isin(['Found', 'Fell'])
]
```

```python
pivot_2005_2009 = pd.pivot_table(
    filtered,
    index='year',
    columns='fall',
    values='mass (g)',
    aggfunc=['count', lambda s: s.quantile(0.95)]
)
```

```python
pivot_2005_2009.columns = [
    f"{'meteorite_count' if stat == 'count' else 'mass_95th_pct_g'}_{fall_type.lower()}"
    for stat, fall_type in pivot_2005_2009.columns
]
```

```python
pivot_2005_2009
```

```python
# found meteor masses
found_meteorites = meteor_landings[meteor_landings['fall']=='Found']
found_meteorites = found_meteorites.sort_index()
found_meteorites
```

```python
fallen_meteorites = meteor_landings[meteor_landings['fall']=='Fell']
fallen_meteorites.sort_index()
fallen_meteorites
```

```python
found_meteor_mean = np.mean(found_meteorites['mass (g)'])
found_meteor_median = np.median(found_meteorites['mass (g)'])
found_meteor_mode = np.array(found_meteorites['mass (g)'].mode())
found_meteor_var = np.var(found_meteorites['mass (g)'])
found_meteor_std = np.std(found_meteorites['mass (g)'])
```

```python
fallen_meteor_mean = np.mean(fallen_meteorites['mass (g)'])
fallen_meteor_median = np.median(fallen_meteorites['mass (g)'])
fallen_meteor_mode = np.array(fallen_meteorites['mass (g)'].mode())
fallen_meteor_var = np.var(fallen_meteorites['mass (g)'])
fallen_meteor_std = np.std(fallen_meteorites['mass (g)'])
```

```python
meteorites_mass_stats = pd.DataFrame({
    'fall_type': ['Found', 'Fell'],
    'mean_mass': [found_meteor_mean, fallen_meteor_mean],
    'median_mass': [found_meteor_median, fallen_meteor_median],
    'mode_mass': [found_meteor_mode, fallen_meteor_mode],
    'variance_mass': [found_meteor_var, fallen_meteor_var],
    'std_mass': [found_meteor_std, fallen_meteor_std]
})
```

```python
meteorites_mass_stats
```

## Exercise Part 5

```python
# Taxi data: hourly totals and top 5 hours by tip amount
taxi = pd.read_csv('../[00] Datasets/2019_Yellow_Taxi_Trip_Data.csv')
```

```python
dropoff_col_candidates = ['tpep_dropoff_datetime', 'dropoff_datetime', 'lpep_dropoff_datetime']
dropoff_col = next((c for c in dropoff_col_candidates if c in taxi.columns), None)
if dropoff_col is None:
    raise KeyError('No dropoff datetime column found in taxi dataset.')
taxi[dropoff_col] = pd.to_datetime(taxi[dropoff_col], errors='coerce')
```

```python
sum_cols = ['trip_distance', 'fare_amount', 'tolls_amount', 'tip_amount']
for col in sum_cols:
    taxi[col] = pd.to_numeric(taxi[col], errors='coerce')
```

```python
hourly_totals = (
    taxi.dropna(subset=[dropoff_col])
        .set_index(dropoff_col)
        .resample('h')[sum_cols]
        .sum()
        .sort_index()
 )
```

```python
top_5_tip_hours = hourly_totals.nlargest(5, 'tip_amount')
top_5_tip_hours
```
