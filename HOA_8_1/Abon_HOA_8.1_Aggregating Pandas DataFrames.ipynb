### Create bins for each full number of magnitude (for example, the first bin is 0-1, the second is 1-2, and so on) with a magType of ml and count how many are in each bin.

```python
ml_df = earthquakes_df[earthquakes_df['magType'] == 'ml'].copy()

max_mag = int(np.ceil(ml_df['mag'].max()))
bins = np.arange(0, max_mag + 1, 1)
labels = [f"{i}-{i+1}" for i in range(0, max_mag)]

ml_df['mag_bin'] = pd.cut(
    ml_df['mag'],
    bins=bins,
    labels=labels,
    right=False,
    include_lowest=True
)

bin_counts = ml_df['mag_bin'].value_counts(sort=False)
bin_counts
```

### Using the faang.csv file, group by the ticker and resample to monthly frequency. Make the following aggregations:
- Mean of the opening price
- Maximum of the high price
- Minimum of the low price

Mean of the closing price

```python
mean_close = faang_df.groupby('ticker', group_keys=True)[['close']].mean()
mean_close
```

Sum of the volume traded

```python
sum_volume = faang_df.groupby('ticker', group_keys=True)[['volume']].sum()
sum_volume
```

### Calculate the rolling 60-day aggregations of OHLC data by ticker for the FAANG data. Use the same aggregations as exercise no. 3.

```python
# Filter stock data from first recorded day to 60 days later
faang_df['date'] = pd.to_datetime(faang_df['date'])
start_date = faang_df['date'].min()
end_date = start_date - pd.Timedelta(days=60)
```

```python
faang_60d = faang_df[(faang_df['date'] > start_date) & (faang_df['date'] <= end_date)]
```

```python
OHLC_60 = faang_df.groupby('ticker').agg(
    open_mean=('open', 'mean'),
    high_max=('high', 'max'),
    low_min=('low', 'min'),
    close_mean=('close', 'mean'),
    volume_sum=('volume', 'sum')
)
OHLC_60
```

### Create a pivot table of the FAANG data that compares the stocks. Put the ticker in the rows and show the averages of the OHLC and volume traded data.

```python
avg_OHLC_volume = faang_df.groupby('ticker').agg(
    open_mean=('open', 'mean'),
    high_mean=('high', 'mean'),
    low_mean=('low', 'mean'),
    close_mean=('close', 'mean'),
    volume_mean=('volume', 'mean')
)
avg_OHLC_volume
```

### Use the transform() method on the FAANG data to represent all the values in terms of the first date in the data. To do so, divide all the values for each ticker by the values for the first date in the data for that ticker. This is referred to as an index, and the data for the first date is the base (https://ec.europa.eu/eurostat/statistics-explained/ index.php/ Beginners:Statisticalconcept-Indexandbaseyear). When data is in this format, we can easily see growth over time. Hint: transform() can take a function name.

```python
faang_index = faang_df.copy()
faang_index['date'] = pd.to_datetime(faang_index['date'])
faang_index = faang_index.sort_values(['ticker', 'date'])

value_cols = ['open', 'high', 'low', 'close', 'volume']
base_values = faang_index.groupby('ticker')[value_cols].transform('first')
faang_index[value_cols] = faang_index[value_cols].div(base_values)

faang_index.groupby('ticker').head(1)
```
