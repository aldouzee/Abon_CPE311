# Tower of Hanoi Problem

### Code

```python
# creation of poles (lists)
A = [10, 8, 5, 3]
B = []
C = []
```

```python
# stage 1 - biggest disk in C, all other disks in B
def stage_1(A, B, C):
        if len(A) == 1:
            C.append(A[0])
            print(f"Move {A[0]} from A to C.")
            A.pop()
        else:
            if len(B) > 1:
                if B[-1] > A[-1]:
                    B.append(A[-1])
                    print(f"Move {A[-1]} from A to B.")
                else:
                    C.append(A[-1])
                    print(f"Move {A[-1]} from A to C.")
            else:
                B.append(A[-1])
                print(f"Move {A[-1]} from A to B.")
            A.pop()

# stage 2- 
def stage_2(A, B, C):
        if len(B) == 1:
            C.append(B[0])
            print(f"Move {B[0]} from B to C.")
            B.pop()
        else:
            if len(C) > 1:
                if C[-1] > B[-1]:
                    C.append(B[-1])
                    print(f"Move {A[-1]} from B to C.")
                else:
                    A.append(B[-1])
                    print(f"Move {B[-1]} from B to A.")
            else:
                A.append(B[-1])
                print(f"Move {A[-1]} from B to A.")
            B.pop()
```

```python
stage_1(A, B, C)
stage_2(A, B, C)
```
